```python
import numpy as np
import cv2
import imutils
import matplotlib.pyplot as plt
from scipy import ndimage


def cacao_processing(image):
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # mewakili warna orange
    low_org = (1, 80, 100)
    upper_org = (10, 255, 255)
    # mewakili warna ungu
    low_ung = (0, 29, 48)
    upper_ung = (14, 143, 133)
    
    mask = cv2.inRange(hsv_image,low_org,upper_org)
    mask2 = cv2.inRange(hsv_image,low_ung,low_ung)
    fin_mask = mask + mask2
    result = cv2.bitwise_and(image.copy(), image.copy(), mask=fin_mask)
    return result, fin_mask

Img = cv2.imread("data1.jpg")
I_r = cv2.resize(Img, None, fx=0.2, fy=0.2)
cv2.imshow("RGB", I_r)
img3,mask = cacao_processing(I_r)
cv2.imshow("Channel", img3)

# SET THRESHOLD
thres = 1
maxValue = 255

# BASIC THRESHOLD
th, dst = cv2.threshold(img3, thres, maxValue, cv2.THRESH_BINARY)
cv2.imshow("Threshold BinaryImage", dst)

# opening
kernelopen = np.ones((3, 1), np.uint8)
opening = cv2.morphologyEx(dst, cv2.MORPH_OPEN, kernelopen)
cv2.imshow("opening", opening)

# closing
kernelclose = np.ones((9,9), np.uint8)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernelclose)
cv2.imshow("closing", closing)
cv2.imwrite("closing.png", closing)

# opening 2
kernelopen2 = np.ones((2,3), np.uint8)
opening2 = cv2.morphologyEx(closing,cv2.MORPH_OPEN,kernelopen)
cv2.imshow("opening2", opening2)
cv2.imwrite("opening2.png", opening2)

Image = cv2.imread("closing.png", cv2.IMREAD_GRAYSCALE)
cv2.imshow("image", Image)
img_2 = Image.copy()

# closing2
kernelclose2 = np.ones((250, 250), np.uint8)
closing2 = cv2.morphologyEx(Image, cv2.MORPH_CLOSE, kernelclose)
cv2.imshow("closing2", closing2)

# Dilasi
kernel = np.ones((3,3),np.uint8)
dilasii = cv2.dilate(closing2,kernel,iterations = 3)
cv2.imshow("dilasi", dilasii)
cv2.imwrite("closing22.png", dilasii)


Image2 = cv2.imread("closing22.png", cv2.IMREAD_GRAYSCALE)
cv2.imshow("image", Image2)


#contours
contours,hierachy=cv2.findContours(Image2,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
jumlah=0
for cnt in contours:
    area = cv2.contourArea(cnt)
        
    if area > 200:
       
        (x, y, w, h) = cv2.boundingRect(cnt)
        cv2.rectangle(I_r, (x, y), (x + w, y + h), (250, 255, 255), 1)

cv2.imshow("contoursbounding", I_r)
cv2.imwrite("contours.jpg", Image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
